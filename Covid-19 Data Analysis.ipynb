```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print("Modules are imported! ")
```

```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
# checking the shape of dataframe (no. of rows and columns)

corona_dataset_csv.shape
```

```python
# deleting useless columns

corona_dataset_csv.drop(['Lat', 'Long'], axis = 1, inplace = True)
```

```python
corona_dataset_csv.head(10)
```

```python
# aggregating the row by the country

corona_dataset_aggregated = corona_dataset_csv.groupby('Country/Region').sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# visualising data related to a country

corona_dataset_aggregated.loc['China'] #return a pandas series 
```

```python
corona_dataset_aggregated.loc['China'].plot()
plt.title("COVID-19 Trend in China")
plt.xlabel("Date")                   # label for x-axis
plt.ylabel("Number of Cases")       #label for y-axis
```

```python
# comparing the rise in cases between 3 counries

corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()



plt.legend()
plt.title("COVID-19 Comparison")
plt.xlabel("Date")                   # label for x-axis
plt.ylabel("Number of Cases")       #label for y-axis
```

```python
# finding a good measure

corona_dataset_aggregated.loc['China'][:3].plot()  
```

```python
# calculating and plotting the first derivative of the curve

corona_dataset_aggregated.loc['China'].diff().plot()

plt.title("Change in infection rates day by day.")
plt.xlabel("Dates")
plt.ylabel("Confirmed Cases")
```

```python
# finding maximum infection rates in china, Italy, Spain

corona_dataset_aggregated.loc['China'].diff().max()
```

```python
corona_dataset_aggregated.loc['Italy'].diff().max()
```

```python
corona_dataset_aggregated.loc['Spain'].diff().max()
```

```python
# finding maximum infection rates of all countries
countries = list(corona_dataset_aggregated.index)
max_infection_rate = []

for c in countries:
    max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max())

corona_dataset_aggregated["max_infection_rate"] = max_infection_rate  
```

```python
corona_dataset_aggregated.head()
```

```python
#creating a new dataframe with only needed column

corona_data = pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
```

```python
corona_data.head()
```

```python
# World Happiness Report

happiness_report_csv = pd.read_csv('Datasets/worldwide_happiness_report.csv')
```

```python
happiness_report_csv.head()
```

```python
useless_cols = ['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
happiness_report_csv.drop(useless_cols, axis = 1, inplace = True)
happiness_report_csv.head()
```

```python
# changing the indices of the datafame

happiness_report_csv.set_index('Country or region', inplace = True)
```

```python
happiness_report_csv.head()
```

```python
corona_data.shape
```

```python
# joining above two datasets

data = corona_data.join(happiness_report_csv, how = "inner")
data.head()
```

```python
# correlation matrix

data.corr()
```

```python
#visualisation of the results

#plotting GDP vs max infection rate
x = data["GDP per capita"]
y = data["max_infection_rate"]

sns.scatterplot(x=x, y=np.log(y))
```

```python
sns.regplot(x=x, y=np.log(y))
```

```python
# Plotting Social support vs max infection rate

x = data["Social support"]
y = data["max_infection_rate"]

sns.scatterplot(x = x, y = np.log(y), color = 'red')
```

```python
sns.regplot(x = x, y = np.log(y), color = 'red')
```

```python
# Plotting Healthy life expectancy vs max infection rate

x = data['Healthy life expectancy']
y = data['max_infection_rate']

sns.scatterplot(x = x, y = np.log(y), color = 'green')
```

```python
sns.regplot(x = x, y = np.log(y), color = 'green')
```
